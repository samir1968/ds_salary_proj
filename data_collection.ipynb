 # Data Cleaning:

### Salary parsing:

```python
df2 = df[df['Salary Estimate'] != '-1']
```

```python
df2.head()
```

```python
df2.shape
```

```python
df.shape[0] - df2.shape[0]
```

```python
(df.shape[0] - df2.shape[0])/df.shape[0] * 100
```

```python
df2["hourly"] = df2['Salary Estimate'].apply(lambda x: 1 if "per hour" in x.lower() else 0)
```

```python
df2.tail(20)
```

```python
df2["employer provided salary"] = df2['Salary Estimate'].apply(lambda x: 1 if "employer provided salary:" in x.lower() else 0)
```

```python
df2.tail(20)
```

```python
salary = df2['Salary Estimate'].apply(lambda x: x.split("(")[0])
```

```python
salary.unique()
```

```python
minus_kd = salary.apply(lambda x: x.replace("K","").replace("$",""))
```

```python
minus_kd.unique()
```

```python
minus_hr = minus_kd.apply(lambda x: x.lower().replace("per hour","").replace("employer provided salary:",""))
```

```python
minus_hr.tail(20)
```

```python
df2["min_salary"] = minus_hr.apply(lambda x: int(x.split("-")[0]))
```

```python
df2.head(3)
```

```python
df2["max_salary"] = minus_hr.apply(lambda x: int(x.split("-")[1]))
```

```python
df2.head(3)
```

```python
df2["avg_salary"] = (df2["min_salary"]+df2["max_salary"])/2
```

```python
df2.head(3)
```

### Company Name Only:

```python
df2["company_txt"] = df2.apply(lambda x: x["Company Name"] if x["Rating"]<0 else x["Company Name"][:-3], axis=1)
```

```python
df2.head(3)
```

### State Field:

```python
df2["job_state"] = df2["Location"].apply(lambda x: x.split(",")[1])
```

```python
df2.head()
```

```python
df2["job_state"].value_counts()
```

```python
df2["same_location"] = df2.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
df2.head(3)
```

### Age Of Company:

```python
df2["age"] = df2.Founded.apply(lambda x: x if x < 1 else 2020 - x)
```

```python
df2.head(3)
```

### Parsing Of Job Description(Python, R-studio, etc...):

### python:

```python
df2["python_yn"] = df2["Job Description"].apply(lambda x: 1 if "python" in x.lower() else 0)
```

```python
df2.head(3)
```

```python
df2["python_yn"].value_counts()
```

### R-Studio

```python
df2["R_yn"] = df2["Job Description"].apply(lambda x: 1 if "r studio" in x.lower() or "r-studio" in x.lower() else 0)
```

```python
df2.head(3)
```

```python
df2["R_yn"].value_counts()
```

### Spark:

```python
df2["spark"] = df2["Job Description"].apply(lambda x: 1 if "spark" in x.lower() else 0)
```

```python
df2.head(3)
```

```python
df2["spark"].value_counts()
```

### AWS:

```python
df2["aws"] = df2["Job Description"].apply(lambda x: 1 if "aws" in x.lower() else 0)
```

```python
df2.head(3)
```

```python
df2["aws"].value_counts()
```

### Excel:

```python
df2["excel"] = df2["Job Description"].apply(lambda x: 1 if "excel" in x.lower() else 0)
```

```python
df2.head(3)
```

```python
df2["excel"].value_counts()
```

### Droping The First Column Of The DataFrame:

```python
df_out = df2.drop("Unnamed: 0", axis=1)
```

```python
df_out.head(3)
```

### Saving The Data To csv File:

```python
df_out.to_csv("salary_data_cleaned.csv", index=False)
```
